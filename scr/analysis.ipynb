```python
from faker import Faker
fake = Faker()
names = [fake.name().split() for i in range (100)]
names = [name for name in names if len(name) == 2]
names
```

```python
import sqlite3
connection = sqlite3.connect("sample.db")
```

```python
create_table_query = '''
CREATE TABLE IF NOT EXISTS people (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL,
    surname TEXT NOT NULL
)
'''
cursor = connection.cursor()
cursor.execute(create_table_query)
connection.commit()
                                                         
```

```python
insert_query = 'INSERT INTO people(name, surname) VALUES(?, ?)'
for name, surname in names:
    cursor.execute(insert_query, (name, surname))
connection.commit()
```

```python
select_query = 'SELECT * FROM people LIMIT 10'
for i in cursor.execute(select_query):
    print(i)
```

```python
query = 'SELECT * FROM people'
for result in cursor.execute(query):
    print(result)
```

```python
query = 'SELECT name, surname FROM people WHERE name LIKE "%A__" LIMIT 100'
for result in cursor.execute(query):
    print(f"First Name: {result[0]}, Surname: {result[1]}")
```
