```python
import os

os.mkdir('sample_data')

files = ['file1.txt', 'file2.csv', 'image.png']
folders = ['reports', 'backups']

for f in files:
    open(os.path.join('sample_data', f), 'w').close()

for folder in folders:
    os.mkdir(os.path.join('sample_data', folder))
```

```python
import os
for item in os.listdir('sample_data'):
    if os.path.isdir(item):
        print(f"This is a directory: {0}".format(item))
    else:
        print(f"This is a file: {0}".format(item))
```
